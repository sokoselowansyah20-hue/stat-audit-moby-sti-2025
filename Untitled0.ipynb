```python
# Data Collection

This notebook collects data from the Pandas GitHub repository using the GitHub REST API.

Repository:
https://github.com/pandas-dev/pandas
```

```python
import requests
import pandas as pd
```

```python
owner = "pandas-dev"
repo = "pandas"
```

```python
url = f"https://api.github.com/repos/{owner}/{repo}/pulls?state=all&per_page=100"

response = requests.get(url)

pull_requests = response.json()

print("Total Pull Requests Retrieved:", len(pull_requests))
```

```python
df_pr = pd.json_normalize(pull_requests)

df_pr.head()
```

```python
print(df_pr.columns.tolist())
```

```python
df_pr = df_pr[
    [
        "id",
        "number",
        "state",
        "created_at",
        "closed_at",
        "merged_at",
        "user.login"
    ]
]

df_pr.head()
```

```python
df_pr.to_csv(
    "pull_requests.csv",
    index=False
)

print("File saved successfully!")
```
