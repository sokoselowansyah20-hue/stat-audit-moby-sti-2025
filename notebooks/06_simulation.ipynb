```python
import numpy as np import matplotlib.pyplot as plt
```

```python
lam = 15 simulations = np.random.poisson( lam=lam, size=10000 ) print("Mean:", simulations.mean())
```

```python
plt.hist(simulations, bins=30) plt.title("Monte Carlo Simulation") plt.show()
```
