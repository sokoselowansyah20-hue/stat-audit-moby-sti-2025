```python
import pandas as pd from scipy.stats import ttest_1samp
```

```python
df = pd.read_csv("../data/raw/pull_requests.csv") df.head()
```

```python
pr_per_day = df.groupby("created_at").size()

pr_per_day.head()
```

```python
test_value = 10 result = ttest_1samp(pr_per_day, test_value) print("Statistic:", result.statistic) print("P-value:", result.pvalue)
```
