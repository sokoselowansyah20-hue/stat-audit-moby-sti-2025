.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("../data/raw/pull_requests.csv")
```

```python
df.info()
```

```python
df["created_at"] = pd.to_datetime(df["created_at"])

monthly = df.groupby(df["created_at"].dt.month).size()

monthly.plot(kind="bar")

plt.title("Pull Requests per Month")

plt.show()
```
