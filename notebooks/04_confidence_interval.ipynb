```python
import pandas as pd import numpy as np from scipy import stats
```

```python
df = pd.read_csv("../data/raw/pull_requests.csv")
```

```python
sample = np.random.randint(1, 100, 100)

mean = np.mean(sample)
std = np.std(sample, ddof=1)

n = len(sample)

ci95 = stats.t.interval(
    confidence=0.95,
    df=n-1,
    loc=mean,
    scale=std/np.sqrt(n)
)

print(ci95)
90%, 95%, and 99% CI
for conf in [0.90, 0.95, 0.99]:

    ci = stats.t.interval(
        confidence=conf,
        df=n-1,
        loc=mean,
        scale=std/np.sqrt(n)
    )

    print(f"{int(conf*100)}% CI:", ci)
```
